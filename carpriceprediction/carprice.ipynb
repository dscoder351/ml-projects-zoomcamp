## 2.1 Loading the data 

```python
#importing libraries
import pandas as pd
import numpy as np
```

```python
#loading the data using url
url = "https://raw.githubusercontent.com/alexeygrigorev/mlbookcamp-code/refs/heads/master/chapter-02-car-price/data.csv"
df = pd.read_csv(url)
df.head()
```

## 2.2 Data Preparation

```python
#df.columns
```

```python
#changing the column to lower case and replacing the space with underscore.
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns
```

```python
df.head()
```

```python
# finding the column data types
df.dtypes
```

```python
#selecting the object data type. Selecting the index column from the series and saving it into a list and storing into variable strings
strings = list(df.dtypes[df.dtypes == 'object'].index)
strings
```

```python
# Normalize the data for object type columns
for col in strings:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```

```python
df.head()
```

#### In the above result all the columns and string values in data are lower case, spaces are replaced with underscores. Dataset is normalized.
